```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import pandas as pd
import sqlite3, sys
```

```python
comandos = sys.argv
nombre_base_datos = comandos[1]
```

```python
#load docs
df = pd.read_excel('PIEZAS_limpias2.xlsx')
df2= pd.read_excel('PIEZAS_sucias2.xlsx')
```

```python
df_validacion_0 = df2[df2['Validacion'] == 0]
df_validacion_1 = df2[df2['Validacion'] == 1]
```

```python
sucias= df_validacion_0['Piezas Sucias'].unique()
limpias= df['Piezas Limpias'].unique()
```

```python
sucias=sucias.astype(str)
limpias=limpias.astype(str)
```

```python
#df_validacion_1.to_excel('Traductor_Piezas.xlsx')
```

```python
partes_auto = []
for x in limpias:
    partes_auto.append(x)
    
for y in sucias:
        print( y, process.extract(y, partes_auto, limit =2, scorer=fuzz.token_set_ratio))
```
