```python
data['KEY']= data['KEY'].astype(str)
```

```python
# Group the filtered data by 'KEY'
grouped_data = data.groupby('KEY')
```

```python
# crear el diccionario year_presence
year_presence = {}

# encontrar el año minmo y maximo
min_year = data['Ano'].min()
max_year = data['Ano'].max()

# Aplicarlo en cada grupo (key)
for key, group in grouped_data:
    # unique years del grupo
    years = group['Ano'].unique()

    # Create a dictionary for the presence of KEY in each year
    year_dict = {year: 1 for year in years}

    # Rellenar los años antes de su primer appearance en la base de datos
    for year in range(min_year, years.min()):
        year_dict[year] = '-'

    # Rellenar los años despues del first appearance
    for year in range(years.min() + 1, max_year + 1):
        if year not in years:
            year_dict[year] = 0

    # Store the year dictionary for the KEY
    year_presence[key] = year_dict

# Create a DataFrame from the year presence dictionary
df = pd.DataFrame(year_presence)

# Sort the columns (years) in ascending order
df = df.reindex(sorted(df.columns), axis=1)

# Transpose el dataframe
df = df.transpose()
```
